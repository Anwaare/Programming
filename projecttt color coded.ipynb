```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

R = 8.314 # J/(mol*K)

def cP(V, n, T):
    return (n*R*T)/V

def cV(P, n, T):
    return (n*R*T)/P

def cT(P, V, n):
    return (P*V)/(n*R)

def cn(P, V, T):
    return (P*V)/(T*R)

def menu():
    while True:
        print("What variable do you want to calculate?")
        print("1. Pressure (P)")
        print("2. Volume (V)")
        print("3. Temperature (T)")
        print("4. n")
        print("5. Exit")
        choice = int(input("Your choice (1-5): "))
        if choice == 1:
            V = float(input("Enter the volume (m^3): "))
            n = float(input("Enter the number of moles of gas(mol^-1): "))
            T = float(input("Enter the temperature (K): "))
            P = cP(V, n, T)
            print("The pressure is ", P, "Pa")
        elif choice == 2:
            P = float(input("Enter the pressure (Pa): "))
            n = float(input("Enter the number of moles of gas: "))
            T = float(input("Enter the temperature (K): "))
            V = cV(P, n, T)
            print("The volume is ", V, "m^3")
        elif choice == 3:
            P = float(input("Enter the pressure (Pa): "))
            V = float(input("Enter the volume (m^3): "))
            n = float(input("Enter the number of moles of gas: "))
            T = cT(P, V, n)
            print("The temperature is ", T, "K")
        elif choice == 4:
            P = float(input("Enter the pressure (Pa): "))
            V = float(input("Enter the volume (m^3): "))
            T = float(input("Enter the temperature (K): "))
            n = cn(P, V, T)
            print("The Number of molas is ", n,"mol^-1")
        elif choice == 5:
            break
        else:
            print("Invalid choice")

menu()
```

```python
import pandas as pd
data1=pd.DataFrame({"Pressure(pa)":[1.004,0.5,0.25,0.11],
                    "Volum(v)":[1000,1928.848,2456,3658],
                    "number of moles(mol^-1)":[3,5.8,1,0.9045475298611144],
                    "temperature(K)":[20,15,10,5]})
result=pd.DataFrame(data1)
result
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.rc('font', family='Times New Roman')
pressure = data1["Pressure(pa)"]
volume = data1["Volum(v)"]
plt.plot(volume, pressure, 'm-')
plt.scatter(volume, pressure, s=100, c='b', marker='o')
plt.xlabel("Volume (m^3)")
plt.ylabel("Pressure (Pa)")
plt.title("Pressure-Volume (PV) Graph")
sns.set()
plt.gca().set_facecolor('#EB7B65') 
plt.show()
```
